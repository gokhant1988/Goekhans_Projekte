```python

# Anwendung RandomForest


#1.Bibliotheken importieren
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error
from sklearn.tree import plot_tree
```

```python
#2. Datensatz laden

from sklearn.datasets import load_iris

iris=load_iris()

df=pd.DataFrame(data=iris.data, columns=iris.feature_names)

df.head()

df["Target"]=iris.target


#3. Erste Ãœbersicht vom Datensatz darstellen 
df.head()
```

```python
#4. Grafische Darstellung mittels Pairplot gruppiert nach "Target"

sns.pairplot(df,hue="Target")
```

```python
# 5. Aufbau der Features anzeigen

iris
```

```python

#6. Wir splitten die Daten in einen Trainings- und Testdatensatz:

X=iris.data
y=iris.target

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)


#7. DecisionTree fitten und predictions berechnen
clf=DecisionTreeClassifier(random_state=0,max_depth=2)

clf.fit(X_train,y_train)

predictions=clf.predict(X_test)

predictions

#8.Accuracy score berechnen und anzeigen

print(accuracy_score(predictions,y_test))
```

```python
import sklearn
sklearn.__version__

#9. Decision Tree grafisch darstellen

viz_tree= plot_tree(clf,feature_names=iris.feature_names, fontsize=8)
```

Ein weiteres Beispiel mit einem Wein-QualitÃ¤t-Datensatz

```python
#1. Datensatz laden

df=pd.read_csv("C:/Users/goex1/Desktop/1 Weiterbildung Data Science/Machine Learning/Woche 6/RandomForest_Anwendung/winequality-white.csv", delimiter=";")

#2. Ãœbersicht vom Datensatz
df.head()

#3. Verteilungsdiagramm darstellen
sns.histplot(data=df, x="quality")
```

```python

#4. Datensatz in Trainings- und Testdaten unterteilen

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(["quality"], axis=1), 
    df["quality"],                
    test_size=0.2, 
    random_state=0
)


#5. Randomforest fitten, anwenden und Accuracy Score berechnen
clf=RandomForestClassifier(max_depth=2, random_state=1, n_estimators=150)

clf.fit(X_train,y_train)

predictions=clf.predict(X_test)

print(accuracy_score(predictions,y_test))

#6. Decision Tree grafisch darstellen

plt.figure(figsize=(15, 15))
plot_tree(clf.estimators_[57],filled=True, feature_names=df.columns,fontsize=11)

plt.show()
```
