```python
import requests
from pprint import pprint

res = requests.get("https://api.imdbapi.dev/titles?types=MOVIE&minAggregateRating=8")

res.text

data=res.json()

data
```

```python
for movie in data["titles"]:
    print(movie["primaryTitle"])
    print("year:", movie["startYear"])
    print(movie["genres"])
    if movie["rating"]["aggregateRating"] is not None:
                           print("aggregate rating:",movie["rating"]["aggregateRating"])
    else: print("Kein rating vorhanden")
```
